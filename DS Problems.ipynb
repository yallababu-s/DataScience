# 7)Test the Analysis of Variance between etest_p and mba_p at signifancelevel 5%.

```python
#7)Test the Analysis of Variance between etest_p and mba_p at signifancelevel 5%.
#(Make decision using Hypothesis Testing)
```

```python
from scipy.stats import f_oneway
etest_p = dataset['etest_p']
mba_p = dataset['mba_p']

# Perform one-way ANOVA  (Analysis of Variance)
f_statistic, p_value = f_oneway(etest_p, mba_p)

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Decision based on significance level (5%)
if p_value < 0.05:
    print("Reject the null hypothesis: There is no significant difference between etest_p and mba_p.")
else:
    print("Fail to reject the null hypothesis: There is a significant difference between etest_p and mba_p.")

#H0=null hypothesis: reject and There is no similarity between them p<0.05
#p-value (2.75 × 10⁻¹⁶) is much smaller
```

# 9)Convert the normal distribution to standard normal distribution for salary column

```python
#9)Convert the normal distribution to standard normal distribution for salary column
#Ans: using zscore  to get mean =0 std=1
```

```python
def stdNDgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

# 13) plot any useful graph and explain it.

```python
#Answer:
#Correlation Heatmap helps identify relationships between numerical features.
#Box Plots provide insight into the distribution of salary for categorical variables, revealing possible disparities.
#Pairplot allows you to see how multiple numerical features relate to salary in a scatterplot matrix.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Visualizing the correlation heatmap
plt.figure(figsize=(12, 8))
corr_matrix = dataset.corr()  # Compute the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Box plot for categorical features vs. salary (e.g., 'gender' and 'workex')
plt.figure(figsize=(12, 6))

# Gender vs Salary
plt.subplot(1, 2, 1)
sns.boxplot(x='gender', y='salary', data=dataset)
plt.title("Gender vs Salary")

# Work experience vs Salary
plt.subplot(1, 2, 2)
sns.boxplot(x='workex', y='salary', data=dataset)
plt.title("Work Experience vs Salary")

plt.tight_layout()
plt.show()

# Pairplot for numerical features (subset of features)
sns.pairplot(dataset[['ssc_p', 'hsc_p', 'degree_p', 'mba_p', 'salary']])
plt.suptitle("Pairplot: Numerical Features vs Salary", y=1.02)
plt.show()
```
