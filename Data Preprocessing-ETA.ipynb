```python
#4)Consider this as a semi supervised, create a model, so that we can predict the missing values.
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual


    
```

```python
quan,qaul=quanQual(dataset)
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
num=imp.transform(dataset[quan])
num = pd.DataFrame(dataset,columns = quan)
```
